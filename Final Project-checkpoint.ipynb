We now overlay the population map on top of a basemap using folium’s choropleth method. 
This method is able to read the geometries and attributes directly from a geodataframe.

```python
%pip install geojson
```

```python
import geopandas as gpd
from pyproj import CRS
import requests
import geojson

# Specify the url for web feature service
url = 'https://kartta.hsy.fi/geoserver/wfs'

# Specify parameters (read data in json format).
# Available feature types in this particular data source: http://geo.stat.fi/geoserver/vaestoruutu/wfs?service=wfs&version=2.0.0&request=describeFeatureType
params = dict(service='WFS',
              version='2.0.0',
              request='GetFeature',
              typeName='asuminen_ja_maankaytto:Vaestotietoruudukko_2020',
              outputFormat='json')

# Fetch data from WFS using requests
r = requests.get(url, params=params)

# Create GeoDataFrame from geojson
data = gpd.GeoDataFrame.from_features(geojson.loads(r.content))
# Clean overall cell
data = data[data['index'] != 27699]
# Check the data
data.head()
```

```python
from pyproj import CRS
# Define crs
data.crs = CRS.from_epsg(3879)
```

Switch the layer to WGS 84 (epsg: 4326)

```python
# Changing to WGS84
data = data.to_crs(epsg=4326)

# Check layer crs definition
print(data.crs)
```

Renaming columns

```python
# Change the name of a column
data = data.rename(columns={'asukkaita': 'pop20'})
```

```python
# Creating a Geo-id which is needed by the Folium - Each row to have a unique identifier for easy identification
data['geoid'] = data.index.astype(str)
```

```python
# Select only needed columns
data = data[['geoid', 'pop20', 'geometry']]

# Convert to geojson (not needed for the simple coropleth map!)
#pop_json = data.to_json()

#check data
data.head()
```

Creating an interactive choropleth map from the population grid:

```python
#Creating a map instance
m = folium.Map(location=[47.8095, 13.0550], tiles = 'cartodbpositron', zoom_start=11, control_scale=True)

# Plot a choropleth map
# Notice: 'geoid' column that we created earlier needs to be assigned always as the first column
folium.Choropleth(
    geo_data=data,
    name='Population in 2020',
    data=data,
    columns=['geoid', 'pop20'],
    key_on='feature.id',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    line_color='white', 
    line_weight=0,
    highlight=False, 
    smooth_factor=1.0,
    #threshold_scale=[100, 250, 500, 1000, 2000],
    legend_name= 'Museums in SALZBURG CITY').add_to(m)

#Show map
m
```
